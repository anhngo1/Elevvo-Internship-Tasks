#Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import classification_report, precision_score, recall_score, f1_score

from imblearn.over_sampling import SMOTE
```

```python
# Load dataset from https://www.kaggle.com/datasets/architsharma01/loan-approval-prediction-dataset/data
df = pd.read_csv("loan_approval_dataset.csv")

# Data Exploration
print(df.head())
print(df.info())
```

#Data Preparation

```python
# Drop missing values if any
data = df.dropna()

# Remove leading spaces
df.columns = df.columns.str.strip()

# Convert from categorical to numeric values
le = LabelEncoder()

for column in df.columns:

    if df[column].dtype == 'object':

        df[column] = le.fit_transform(df[column])
```

```python
# Target and feature variables
X, y = df.drop('loan_status', axis=1), df['loan_status']

#Split and test data
x_train, x_test, y_train, y_test = train_test_split(X,y,random_state=42)

# Use SMOTE to balance the training dataset
smote = SMOTE(random_state=42)

x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)

print("Original class distribution:")
print(y_train.value_counts())

print("\nBalanced class distribution:")
print(pd.Series(y_train_smote).value_counts())
```

# Models

```python
# Logistic Regression

log_reg = LogisticRegression(max_iter=1000).fit(x_train_smote, y_train_smote)

y_pred_log = log_reg.predict(x_test)

precision_log = precision_score(y_test, y_pred_log)
recall_log = recall_score(y_test, y_pred_log)
f1_log = f1_score(y_test, y_pred_log)

print("Logistic Regression Precision:", round(precision_log,4))
print("Logistic Regression Recall:", round(recall_log,4))
print("Logistic Regression F1 Score:", round(f1_log,4))
```

```python
# Decision Tree

tree = DecisionTreeClassifier(random_state=42).fit(x_train_smote, y_train_smote)

y_pred_tree = tree.predict(x_test)

precision_tree = precision_score(y_test, y_pred_tree)
recall_tree = recall_score(y_test, y_pred_tree)
f1_tree = f1_score(y_test, y_pred_tree)

print("Decision Tree Precision:", round(precision_tree,4))
print("Decision Tree Recall:", round(recall_tree,4))
print("Decision Tree F1 Score:", round(f1_tree,4))
```
